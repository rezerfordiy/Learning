```python
class Subject:
  def __init__(self, name, score):
    self.name = name
    self.score = score

  def learn(self):
    print(f"Предмет {self.name} c оценкой {self.score}")

class Student:
  def __init__(self, name, age):
    self.subjects= []
    self.name = name
    self.age = age

  def add_subject(self, subj):
    self.subjects.append(subj)

  def learn(self):
    for sub in self.subjects:
      print(f"student {self.name} учит")
      sub.learn()


sub1 = Subject("rus", 59)
sub2 = Subject("math", 70)
s = Student("Arkadiy", 18)
s.add_subject(sub1)
s.add_subject(sub2)
s.learn()
```

```python
from abc import ABC, abstractmethod
class AudioDevice(ABC):
  @abstractmethod
  def playSound(self):
    pass

class Headphones(AudioDevice):

  def playSound(self):
    print("Headphones are playing")


class Speakers(AudioDevice):
  def playSound(self):
    print("Speakers are playing")



h = Headphones()
h.playSound()
s = Speakers()
s.playSound()
```

```python
from abc import ABC, abstractmethod

class IPayment():
  @abstractmethod
  def updateAmount(self, amount):
    pass
  @abstractmethod
  def getPaymentDetails(self):
    pass

class Payment(IPayment):
  def __init__(self, amount, date):
    self.amount = amount
    self.date = date

  def updateAmount(self, amount):
    if amount >= 0:
      self.amount = amount
    else:
      print("wrong")

  def getPaymentDetails(self):
    print(f"Payment {self.date} with amount = {self.amount}")


p = Payment(123, "12.12.1222")

p.getPaymentDetails()
p.updateAmount(0)
p.getPaymentDetails()


```
