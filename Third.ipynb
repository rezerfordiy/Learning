```python
n = int(input("enter N: "))
m = int(input("enter M: "))
mas = []
for i in range(n):
  mas.append([0] * m)
for i in range(n):
  for j in range(m):
    mas[i][j] = int(input())


def gen():
  yield 0
  yield 1
  # я думаю что надо дважды счиать 1 тк фибоначи 0 1 1 2 3 5 8...
  yield 1
  a = 1
  b = 1


  while True:
    c = a + b
    b = a
    a = c
    yield c

help = gen()
# будем считать что числа нумеруется построчкам
'''
для этого примера
0 1 2
3 4 5


в принципе
0 1 2 3 4 5 ....
m m+1 ....
2m
.
.
.
'''
sum = 0
for i in range(n * m):
  a = next(help)
  x = a // m
  y = a % m
  if x < n and y < m:
    sum += mas[x][y]
  else:
    break

print(sum)
```

```python
d = {
    "a":(1, 2),
    "b":(1, 2),
    "c":1
}



def f(d):
  nn = {}
  for k,v in d.items():
    if v in nn:
      if isinstance(nn[v], list):
        nn[v].append(k)
      else:
        nn[v] = [nn[v], k]
    else:
      nn[v] = k
  return nn

print(f(d))

```

```python
d1 = {
    "a":1,
    "b":2,
    "c":1
}
d2 = {"a":2, "b": -2, "g":1111}


def f(d1, d2):
  d = d1.copy()
  for k,v in d2.items():
    if k in d:
      d[k] += v
    else:
      d[k] = v

  return d

print(d1, d2, f(d1, d2), sep='\n')
```
