```python
def gcd(a, b):
  while b > 0:
    r = a % b
    a = b
    b = r
  return a

def sm(p1,q1,p2,q2):
  p = p1 * q2 + p2 * q1
  q = q1 * q2
  z = gcd(p, q)
  p /= z
  q /= z
  return p,q



a1 = 1
sum = 1
sumold = 0
eps = 0.001
p,q = 1,1
i = 2
while abs(sum - sumold) > eps:
  sumold = sum

  p,q = sm(p, q, -1 if i % 2 == 0 else 1, i)


  sum = p / q
  i+=1

print(p, q, p/q)

```
