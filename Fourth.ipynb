```python
s1 = input()
s2 = input()
a = list(set(s1.split()).intersection(set(s2.split())))
a.sort(key=lambda x: len(x))
i = len(a) - 1
l = len(a[-1])
while i >= 0 and len(a[i]) == l:
  i-=1
print(a[i+1:])
```

```python
s = 'input.txt'
f = open(s, 'r')
s = f.readlines()
f.close()
for i in range(len(s)):
  s[i] = s[i].strip()
s.sort(key=lambda x:len(x))
l = len(s[0])
i = 0
while i < len(s) and l == len(s[i]):
  i+=1




f = open('output.txt', 'w')
for i in s:
  f.write(i + '\n')
f.close()
```

```python
s = 'input.txt'
f = open(s, 'r')
s = list(map(int, f.readlines()))
f.close()

s = s[len(s) - 50:]
f = open('output.txt', 'w')
for i in s:
  f.write(str(i) + '\n')
f.close()
```
